```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_rnn = pd.read_csv("forecast_power_rnn.csv")
df_lstm = pd.read_csv("forecast_power_lstm.csv")
df_tnn = pd.read_csv("forecast_power_tnn.csv")

mases, smapes, mses = [], [], []
mase_list = df_rnn.iloc[:,1].tolist()
mases.append(np.array([float(value.split(' ')[len(value.split(' ')) -1 ].split('}')[0]) for value in mase_list]))
smape_list = df_rnn.iloc[:,2].tolist()
smapes.append(np.array([float(value.split(' ')[len(value.split(' ')) -1 ].split('}')[0]) for value in smape_list]))
mse_list = df_rnn.iloc[:,3].tolist()
mses.append(np.array(mse_list))

mase_list = df_lstm.iloc[:,1].tolist()
mases.append(np.array([float(value.split(' ')[len(value.split(' ')) -1 ].split('}')[0]) for value in mase_list]))
smape_list = df_lstm.iloc[:,2].tolist()
smapes.append(np.array([float(value.split(' ')[len(value.split(' ')) -1 ].split('}')[0]) for value in smape_list]))
mse_list = df_lstm.iloc[:,3].tolist()
mses.append(np.array(mse_list))

mase_list = df_tnn.iloc[:,1].tolist()
mases.append(np.array([float(value.split(' ')[len(value.split(' ')) -1 ].split('}')[0]) for value in mase_list]))
smape_list = df_tnn.iloc[:,2].tolist()
smapes.append(np.array([float(value.split(' ')[len(value.split(' ')) -1 ].split('}')[0]) for value in smape_list]))
mse_list = df_tnn.iloc[:,3].tolist()
mses.append(np.array(mse_list))

days = range(1,len(mases[0])+1)

plt.plot(days, mases[0], label="RNN")
plt.plot(days, mases[1][:len(mases[0])], label="LSTM")
plt.plot(days, mases[2], label="Transformer")
plt.title("MASE")
plt.xlabel('Length of prediction period in days')
plt.legend()
plt.show()

plt.plot(days, smapes[0], label="RNN")
plt.plot(days, smapes[1][:len(mases[0])], label="LSTM")
plt.plot(days, smapes[2], label="Transformer")
plt.title("SMAPE")
plt.xlabel('Length of prediction period in days')
plt.legend()
plt.show()

plt.plot(days, mses[0], label="RNN")
plt.plot(days, mses[1][:len(mases[0])], label="LSTM")
plt.plot(days, mses[2], label="Transformer")
plt.title("MSE")
plt.xlabel('Length of prediction period in days')
plt.legend()
plt.show()
```
